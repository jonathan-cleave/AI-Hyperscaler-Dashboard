# Comparable Firms by Total

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(similar_df)
X_scaled_df = pd.DataFrame(X_scaled, index=firms_df.index)

X_scaled_df = X_scaled_df.fillna(0)
```

```python
# Find Microsoft row
ticker = "MSFT"
msft_idx = firms_df.index[firms_df["Exchange:Ticker"] == ticker][0]
msft_pos = firms_df.index.get_loc(msft_idx)

  # Distance from Microsoft to every firm
distances = np.linalg.norm(X_scaled_df - X_scaled_df.loc[msft_pos], axis=1)

  # Add distances back to firms_df
closest = firms_df.copy()
closest[f"Distance_to_{ticker}"] = distances

  # Top 4 closest firms, excluding Microsoft itself
top_4_comps = (
      closest[closest["Exchange:Ticker"] != ticker]
      .sort_values(f"Distance_to_{ticker}")
      .head(4)
  )

top_4_comps[
      ["Company Name", "Exchange:Ticker", "Industry Group", "Primary Sector", "Country", f"Distance_to_{ticker}"]
]
```

# Comparable Firms by SIC/Industry/Sector

```python
# Choose how you want to filter comparable firms:
# Options: "sic", "industry", "sector", or "all", "none"
compare_by = "none"
target_ticker = "MSFT"
n_comps = 4

  # Find Microsoft
target_idx = firms_df.index[firms_df["Exchange:Ticker"] == target_ticker][0]
target_row = firms_df.loc[target_idx]

  # Create numeric comparison dataset
  # This follows your approach: remove company descriptor columns
similar_df = firms_df.copy()
similar_df = similar_df.drop(similar_df.columns[0:8], axis=1)

similar_df = similar_df.replace([np.inf, -np.inf], np.nan)
similar_df = similar_df.apply(pd.to_numeric, errors="coerce")

  # Important: remove columns where every value is missing
similar_df = similar_df.dropna(axis=1, how="all")

similar_df = similar_df.fillna(similar_df.median(numeric_only=True))

scaler = StandardScaler()
X_scaled = scaler.fit_transform(similar_df)

X_scaled_df = pd.DataFrame(
      X_scaled,
      index=firms_df.index,
      columns=similar_df.columns
  ).fillna(0)

  # Pick comparable universe
if compare_by == "sic":
      mask = firms_df["SIC Code"] == target_row["SIC Code"]

elif compare_by == "industry":
      mask = firms_df["Industry Group"] == target_row["Industry Group"]

elif compare_by == "sector":
      mask = firms_df["Primary Sector"] == target_row["Primary Sector"]

elif compare_by == "all":
    mask = (
        (firms_df["SIC Code"] == target_row["SIC Code"]) &
        (firms_df["Industry Group"] == target_row["Industry Group"]) &
        (firms_df["Primary Sector"] == target_row["Primary Sector"])
    )

elif compare_by == "none":
    mask = pd.Series(True, index=firms_df.index)

else:
      raise ValueError("compare_by must be 'sic', 'industry', 'sector', or 'all'")

  # Keep only filtered comparable firms
filtered_df = firms_df[mask].copy()

  # Use filtered_df's original index labels to pull scaled rows
candidate_idx = filtered_df.index

distances = np.linalg.norm(
      X_scaled_df.loc[candidate_idx].values - X_scaled_df.loc[target_idx].values,
      axis=1
  )

filtered_df[f"Distance_to_{target_ticker}"] = distances

  # Get top comparable firms, excluding Microsoft
top_comps = (
      filtered_df[filtered_df["Exchange:Ticker"] != target_ticker]
      .sort_values(f"Distance_to_{target_ticker}")
      .head(n_comps)
)

top_comps[
      [
          "Company Name",
          "Exchange:Ticker",
          "SIC Code",
          "Industry Group",
          "Primary Sector",
          "Country",
          f"Distance_to_{target_ticker}"
      ]
  ]
```
